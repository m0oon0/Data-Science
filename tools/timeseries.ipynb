분석을 위한 함수

```python
df['Close_lag'] = df['Close'].shift(periods=1))
df.head()
```

```python
# 백분율 변화량 
df['pct_change'] = df['Close'].pct_change(periods=1)
df.head()
```

```python
df['Close_diff'] = df['Close'].diff(periods=1)
df.head()
```

이동 평균선

```python
df['MA_5'] = df['Close'].rolling(window=5).mean()
df.head()
```

시간 간격 재조정

```python
index= pd.date_range(start='2021-01-01', end = '2021-09-30', freq='B')
series = pd.Series(range(len(index)), index = index)
series
```

```python
series.resample(rule='M').sum()
```

```python
series.resample(rule='M').first()
```

```python
series.resample(rule='B').first() # 영업일만
```
